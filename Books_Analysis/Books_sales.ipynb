```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Books_Data_clean.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df = df[df['Publishing Year'] > 1900]
```

```python
df.isna().sum()
```

```python
df.dropna(subset = "Book Name", inplace = True)
```

```python
df.duplicated().sum()
```

```python
df.nunique()
```

```python
plt.hist(df['Publishing Year'])
plt.title('Distribution of Publishing Years')
plt.xlabel('Publishing Year')
plt.ylabel('Frequency')
plt.show()
```

```python
df['genre'].value_counts().plot(kind='bar')
plt.title('Number of Books per Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Books')
plt.show()
```

```python
df.groupby('Author')['Book_average_rating'].mean().sort_values(ascending=False).head(10)
```

```python
sns.boxplot(x='genre', y='Book_ratings_count', data=df)
plt.title('Book Ratings Count by Genre')
plt.xlabel('Genre')
plt.ylabel('Book Ratings Count')
plt.show()
```

```python
plt.scatter(df['sale price'], df['units sold'])
plt.title('Units Sold vs Sale Price')
plt.xlabel('Sale Price')
plt.ylabel('Units Sold')
plt.show()
```

```python
language_counts = df['language_code'].value_counts()
```

```python
plt.pie(language_counts, labels=language_counts.index, startangle =90, autopct='%1.1f%%')
plt.title('language Distribution of Books')
plt.show()
```

```python
plt.scatter(df['Book_average_rating'], df['Book_ratings_count'])
plt.title('Book Ratings Count vs Average Rating')
plt.xlabel('Book Average Rating')
plt.ylabel('Book Ratings Count')
plt.show()
```

```python
total_gros_sales_by_author = df.groupby('Author')['gross sales'].sum().sort_values(ascending=False).head(10)
```

```python
total_gros_sales_by_author.sort_values(ascending=False).head(10).plot(kind='bar')
plt.title('Top 10 Authors by Gross Sales')
plt.xlabel('Author')
plt.ylabel('Total Gross Sales')
plt.show()
```

```python
sns.boxplot(x='genre', y='units sold', data=df)
plt.title('Units Sold by Genre')
plt.xlabel('Genre')
plt.ylabel('Units Sold')
plt.show()
```

```python
df.groupby('Publishing Year')['units sold'].mean().plot(kind = 'line', marker = 'o')
plt.title('Average Units Sold by Publishing Year')
plt.xlabel('Publishing Year')
plt.ylabel('Average Units Sold')
plt.show()
```
